```python
#Hacer web scrapping con requests y BeautifulSoup

import requests
from bs4 import BeautifulSoup

url = 'https://www.geeksforgeeks.org/html-tables/'
url2 = 'https://app.btgpactual.cl/mercados-en-linea'

page = requests.get(url)
page2 = requests.get(url2,headers={'User-Agent': 'Mozilla/5.0'})

soup = BeautifulSoup(page.content, 'html.parser')
soup2 = BeautifulSoup(page2.content, 'html.parser')

#print(soup.prettify())

#Obtener las tablas de la página
tables = soup.find_all('table')
tables2 = soup2.find_all('table')

#convertir las tablas en un DataFrame
import pandas as pd
df = pd.read_html(str(tables))

df

```

```python
#Bajar archivo de la web y guardar en un archivo

url = 'https://archive.ics.uci.edu/static/public/53/iris.zip'
r = requests.get(url)

with open('iris.zip', 'wb') as f:
    f.write(r.content)

#Descomprimir el archivo
import zipfile
with zipfile.ZipFile('iris.zip', 'r') as zip_ref:
    zip_ref.extractall('iris')

#Leer el archivo descomprimido iris.data y convertirlo en dataframe
df = pd.read_csv('iris/iris.data', header=None)
df
```

```python
#leer precios de vuelos de la web de latam.com con selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import time


url = 'https://www.latamairlines.com/cl/es/ofertas-vuelos?origin=SCL&inbound=2024-07-24T12%3A00%3A00.000Z&outbound=2024-06-20T12%3A00%3A00.000Z&destination=MAD&adt=1&chd=0&inf=0&trip=RT&cabin=Economy&redemption=false&sort=RECOMMENDED'

driver = webdriver.Chrome()
driver.get(url)
time.sleep(5)

#hacer click en span para ver los precios
elemento=driver.find_element(By.ID,"FlightInfoComponent1")

for a in elemento.find_elements(By.TAG_NAME, 'div'):
    print(a.text)

```
